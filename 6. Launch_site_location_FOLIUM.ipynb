```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
import numpy as np
from folium import DivIcon

# Define function to calculate distances using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    
    a = np.sin(delta_phi / 2.0)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2.0)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    return R * c  # Distance in km

# Example coordinates (replace with actual closest locations)
closest_city = [28.40, -80.60]  # Sample city
closest_railway = [28.50, -80.62]  # Sample railway
closest_highway = [28.55, -80.58]  # Sample highway

# Launch site
launch_site = [28.573255, -80.646895]

# Calculate distances
city_distance = haversine(launch_site[0], launch_site[1], closest_city[0], closest_city[1])
railway_distance = haversine(launch_site[0], launch_site[1], closest_railway[0], closest_railway[1])
highway_distance = haversine(launch_site[0], launch_site[1], closest_highway[0], closest_highway[1])

# Create a folium map centered at the launch site
site_map = folium.Map(location=launch_site, zoom_start=10)

# Function to add distance markers
def add_distance_marker(location, distance, label):
    folium.Marker(
        location=location,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM - {}</b></div>'.format(distance, label),
        )
    ).add_to(site_map)

# Add markers for each location with distances
add_distance_marker(closest_city, city_distance, "Closest City")
add_distance_marker(closest_railway, railway_distance, "Closest Railway")
add_distance_marker(closest_highway, highway_distance, "Closest Highway")

# Draw lines from launch site to each location
folium.PolyLine([launch_site, closest_city], color="blue", weight=2.5).add_to(site_map)
folium.PolyLine([launch_site, closest_railway], color="red", weight=2.5).add_to(site_map)
folium.PolyLine([launch_site, closest_highway], color="green", weight=2.5).add_to(site_map)

# Display the map
site_map
```
